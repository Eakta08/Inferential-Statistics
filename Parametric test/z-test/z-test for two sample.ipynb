# <u>z-test for two sample

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

***
**Random samples of 75 males and 50 female’s donors yields mean concentration of 28 and 33 ppm respectively. The amount of trace elements in blood varies with the standard deviation 14.1 and 9.5 ppm respectively for males and females. What is the likelihood that the population means of concentration of elements are the same for men and women.**

_Let alpha value is 0.05 with corresponding Z score is 1.96_

**`Step 1. State the hypotheses. `**

<font face='Comic Sans MS'>The null hypothesis (H0): population means of concentration of elements are the same for men and women

<font face='Comic Sans MS'>The alternative hypothesis (H1): population means of concentration of elements are not the same for men and women (two-tailed test)

**`Step 2. Perform z-test statistics`**

```python
def z_test(xbar,ybar,diff_mux_muy,s1,s2,n1,n2):
    num=((xbar-ybar)-(diff_mux_muy))
    den=np.sqrt(((s1**2)/n1)+((s2**2)/n2))
    return abs(num/round(den,2))
```

```python
sample_mean_a=33
```

```python
sample_mean_b=28
```

```python
n1=75
```

```python
n2=50
```

```python
diff_population_mean_A_and_B=0
```

```python
std_a=14.1
```

```python
std_b=9.5
```

```python
z_test(sample_mean_a,sample_mean_b,diff_population_mean_A_and_B,round(std_a,2),round(std_b,2),n1,n2)
```

```python
critical_value=1.96
```

**`Step 3. Conclusion`**


<font face='Comic Sans MS'>As calculated z > critical value of z, we may Reject H0 at level of significance=0.05. Therefore we conclude that population means of concentration of elements are not the same for men and women.
