# <u>t-test for Difference of Means

***
**Below are given the gain in weights (in lbs.) of pigs fed on two diets A an B.** 

**_Gain in weight:_**

**Diet A : 25,32,30,34,24,14,32,24,30,31,35,25**

**Diet-B : 44,34,22,10,47,31.40,30,32,35,18,21,35,29,22**

**Test if the two diets differ significantly as regards their effect on increase in weight.**


**`Step 1. State the hypotheses. `**

<font face='Comic Sans MS'>The null hypothesis (H0): two diets doesnot differ significantly i.e. population mean of diet A = population mean of diet B

<font face='Comic Sans MS'>The alternative hypothesis (H1): two diets differ significantly i.e. population mean of diet A is not equal to population mean of diet B (two-tailed test)

**`Step 2. Perform t-test statistics`**

```python
import numpy as np
```

```python
def t_test(xbar,ybar,diff_mux_muy,n1,n2):
    s2=((n1*(x.std()**2))+(n2*(y.std()**2)))/(n1+n2-2)
    num=((xbar-ybar)-(diff_mux_muy))
    den=np.sqrt((s2)*((n1+n2)/(n1*n2)))
    return abs(num/den)
```

```python
x=np.array([25,32,30,34,24,14,32,24,30,31,35,25])
y=np.array([44,34,22,10,47,31,40,30,32,35,18,21,35,29,22])
```

```python
import matplotlib.pyplot as plt
plt.plot(x,color='green',marker='o')
plt.plot(y,color='navy',marker='o')
plt.legend(["Diet A",'Diet B'])
plt.show()
```

```python
sample_mean_a=x.mean()
sample_mean_a
```

```python
sample_mean_b=y.mean()
sample_mean_b
```

```python
n1=len(x)
n1
```

```python
n2=len(y)
n2
```

```python
diff_population_mean_A_and_B=0
```

```python
t_test(round(sample_mean_a,2),round(sample_mean_b,2),diff_population_mean_A_and_B,n1,n2)
```

```python
critical_value=2.06
```

**`Step 3. Conclusion`**


<font face='Comic Sans MS'>As calculated t < critical value of t, we may Accept H0 at level of significance=0.05. Therefore we conclude that the two diets doesnot differ significantly on increase in weight.

**Two independent groups of 10 children were tested to find how many digits they could repeat from memory after hearing them. The results are as follows:**

**Group A : 8, 6, 5, 7, 6, 8, 7, 4, 5, 6** 

**Group B : 1, 0, 6, 7, 8, 6, 9, 7, 6, 7, 7**

**Is the difference between the mean scores·of the two groups significant ?**

**`Step 1. State the hypotheses. `**

<font face='Comic Sans MS'>The null hypothesis (H0): two mean scores doesnot differ significantly i.e. population mean of group A = population mean of group B

<font face='Comic Sans MS'>The alternative hypothesis (H1): two mean scores differ significantly i.e. population mean of group A is not equal to population mean of group B

**`Step 2. Perform t-test statistics`**

```python
a=np.array([8, 6, 5, 7, 6, 8, 7, 4, 5, 6])
b=np.array([1, 0, 6, 7, 8, 6, 9, 7, 6, 7, 7])
```

```python
import matplotlib.pyplot as plt
plt.plot(a,color='green',marker='o')
plt.plot(b,color='navy',marker='o')
plt.legend(["Group A",'Group B'])
plt.show()
```

```python
sample_mean_a=a.mean()
sample_mean_a
```

```python
sample_mean_b=b.mean()
sample_mean_b
```

```python
n1=len(a)
n1
```

```python
n2=len(b)
n2
```

```python
diff_population_mean_A_and_B=0
```

```python
t_test(round(sample_mean_a,2),round(sample_mean_b,2),diff_population_mean_A_and_B,n1,n2)
```

```python
critical_value=2.093
```

**`Step 3. Conclusion`**


<font face='Comic Sans MS'>As calculated t < critical value of t, we may Accept H0 at level of significance=0.05. Therefore we conclude that the two mean scores doesnot differ significantly.
