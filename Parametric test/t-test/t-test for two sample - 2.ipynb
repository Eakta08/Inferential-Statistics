# <u>Paired t-test For Difference of Means

***
**A certain stimulus administered to each of the 12 patients resulted in the following increase of blood pressure: 5, 2, 8, -1, 3, 0, -2, 1, 5, 0, 4 and 6. Can it be concluded that the stimulus will, in general. be accompanied by an increase in blood pressure?**

_(Tabulated t at 0.05 level of significance for 11 d.f. for right-tail test is 180)_

**`Step 1. State the hypotheses. `**

<font face='Comic Sans MS'>Null Hypothesis (H0) : there is no significant difference in the blood pressure readings of the patients before and after the drug. In other words, the given increments are just by chance (fluctuations of sampling) and not due to the stimulus.

<font face='Comic Sans MS'>Alternative Hypothesis (H1) : the stimulus results is an increase in blood pressure. (right-tailed test)

**`Step 2. Perform t-test statistics`**

```python
import numpy as np
import pandas as pd
```

```python
def stand_dev_sq(n,d):
    sig_d_sq_by_n= (data.d.sum()**2)/n
    result= ((data['d^2'].sum())-(sig_d_sq_by_n))/(n-1)
    return round(result,4)
def t_test(d,n):
    dbar=(data.d.mean())
    t=dbar/(np.sqrt(stand_dev_sq(n,d)/n))
    return abs(t)
```

```python
data=pd.DataFrame({'d':[5,2,8,-1,3,0,-2,1,5,0,4,6]})
```

```python
data['d^2']=data.d**2
```

```python
data
```

```python
t_test(data.d,len(data.d))
```

```python
critical_value=1.80
```

**`Step 3. Conclusion`**


<font face='Comic Sans MS'>Since calculated t > critical value of t, Ho is rejected at 5% level of significance. Hence we conclude that the stimulus will, in general, be accompanied by an increase in blood pressure.

**In a certain experiment to compare two types of pig foods A and B. the following results of increase in weights were observed in pigs:**



**Assuming that the same set of eight pigs were used in both the foods, can we conclude that food B is better than food A?**

_(Tabulated t at 0.05 level of significance for 7 df for one-tail test is 190)_

**`Step 1. State the hypotheses. `**

<font face='Comic Sans MS'>Null Hypothesis (H0) : there is no significant difference in increase in weights due to food A and B.
    
<font face='Comic Sans MS'>Alternative Hypothesis (H1) : food B is better than food A (Left-tailed). 

**`Step 2. Perform t-test statistics`**

```python
data=pd.DataFrame({'X':[49,53,51,52,47,50,52,53],'Y':[52,55,52,53,50,54,54,53]})
```

```python
data['d']=data.X-data.Y
```

```python
data['d^2']=data.d**2
```

```python
data
```

```python
import matplotlib.pyplot as plt
plt.plot(data.X,marker='o')
plt.plot(data.Y,marker='.')
plt.legend(['Food A','Food B'])
plt.show()
```

```python
t_test(data.d,len(data.d))
```

```python
critical_value=1.90
```

**`Step 3. Conclusion`**


<font face='Comic Sans MS'>Since calculated t > critical value of t, Ho is rejected at 5% level of significance. We conclude that food B is better than food A.
