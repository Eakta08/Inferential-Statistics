# Independent Samples t-test



```python
import numpy as np
from scipy.stats import ttest_ind as ttest
import matplotlib.pyplot as plt
```

***
**Below are given the gain in weights (in lbs.) of pigs fed on two diets A an B.** 

**_Gain in weight:_**

**Diet A : 25,32,30,34,24,14,32,24,30,31,35,25**

**Diet-B : 44,34,22,10,47,31.40,30,32,35,18,21,35,29,22**

**Test if the two diets differ significantly as regards their effect on increase in weight.**

**`Step 1. State the hypotheses. `**

<font face='Comic Sans MS'>The null hypothesis (H0): two diets doesnot differ significantly i.e. population mean of diet A = population mean of diet B

<font face='Comic Sans MS'>The alternative hypothesis (H1): two diets differ significantly i.e. population mean of diet A is not equal to population mean of diet B (two-tailed test)
    
**`Step 2. Perform t-test statistics`**

```python
a=np.array([25,32,30,34,24,14,32,24,30,31,35,25])
b=np.array([44,34,22,10,47,31,40,30,32,35,18,21,35,29,22])
```

```python
plt.plot(a,color='green',marker='o')
plt.plot(b,color='navy',marker='o')
plt.legend(["Diet A",'Diet B'])
plt.show()
```

```python
ttest(a,b,alternative='two-sided')
```

**`Step 3. Conclusion`**


<font face='Comic Sans MS'>As p-value(0.5471) > alpha (0.05), we may Accept H0 at level of significance=0.05. Therefore we conclude that the two diets doesnot differ significantly on increase in weight.
