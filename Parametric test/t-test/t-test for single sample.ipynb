# <u>t-Test for Single Mean</u>


***


**A time study engineer developed a new sequence of operation elements that he hopes will reduce the mean cycle time of a certain production process. The results of a time study of 20 cycles are given below: _(cycle time in minutes)_ :12·25, 11·97, 12·15, 12·08, 12·31, 12·28, 11.94, 11·89, 12·16, 12·04, 12·09, 12·15, 12·14, 12·47, 11·98, 12·04, 12·11, 12·25, 12·15, 12·34. If the present mean cycle time is 12·5 minutes. should he adopt the new sequence?**
*(The critical value of t-test at 0.05 level of significance with 19 degrees of freedom is 2.09302406)*

**`Step 1. State the hypotheses. `**

<font face='Comic Sans MS'>The null hypothesis (H0): mean cycle time = 12.5 minutes

<font face='Comic Sans MS'>The alternative hypothesis (H1): mean cycle time < 12.5 minutes (Left-tailed test)

**`Step 2. Perform t-test statistics`**

```python
import numpy as np
```

```python
def t_test(xbar,mu,s,n):
    t=(xbar-mu)/(np.sqrt((s**2)/n))
    return abs(t)
```

```python
def confidence_interval(xbar,critical_value,s,n):
    a=xbar-((critical_value*s)/np.sqrt(n))
    b=xbar+((critical_value*s)/np.sqrt(n))
    return [a,b]
```

```python
data=np.array([12.25,11.97,12.15,12.08,12.31,12.28,11.94,
              11.89,12.16,12.04,12.09,12.15,12.14,12.47,11.98,
              12.04,12.11,12.25,12.15,12.34])
```

```python
import matplotlib.pyplot as plt
plt.plot(data,color='green',marker='o')
plt.legend(['mean cycle time'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(data=data,orient='v',color='pink') 
sns.swarmplot(data=data,color='b',size=4.5,orient='v')
plt.show()
```

```python
sample_mean=np.mean(data)
sample_mean
```

```python
sample_std=np.std(data)
sample_std
```

```python
n=len(data)
n
```

```python
population_mean = 12.5
```

```python
t_test(sample_mean,population_mean,sample_std,n)
```

```python
critical_value=1.72913282
```

```python
confidence_interval(sample_mean,critical_value,sample_std,n)
```

**` Step 3. Conclusion`**

<font face='Comic Sans MS'>As calculated t > critical value of t, we may Reject H0 at level of significance=0.05. We may conclude that mean cycle time is less than 12.5 minutes.

**A random sample of 10 boys had the following I.Q.' s : 70, 120, 110, 101, 88, 83, 95, 98, 107, 100. Do' these data support the assumption of a population mean I.Q. of 100 ? Find a reasonable range in which most of the mean I.Q. values of samples of 10 boys lie.**
_(Tabulated t-test value for 9 d.f for two-tailed test is 2·262.)_

**`Step 1. State the hypotheses. `**

<font face='Comic Sans MS'>The null hypothesis (H0): population mean I.Q. is equal to 100

<font face='Comic Sans MS'>The alternative hypothesis (H1): population mean I.Q. is not equal to 100

**`Step 2. Perform t-test statistics`**

```python
data=np.array([70,120,110,101,88,83,95,98,107,100])
```

```python
import matplotlib.pyplot as plt
plt.plot(data,color='blue',marker='o')
plt.legend(['I.Q. of boys'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(data=data,orient='v') 
sns.swarmplot(data=data,color='r',size=4.5,orient='v')
plt.show()
```

```python
sample_mean=np.mean(data)
sample_mean
```

```python
population_mean = 100
```

```python
n=len(data)
n
```

```python
sample_std=np.std(data)
sample_std
```

```python
t_test(sample_mean,population_mean,sample_std,n)
```

```python
critical_value=2.262
```

```python
confidence_interval(sample_mean,critical_value,sample_std,n)
```

**` Step 3. Conclusion`**

<font face='Comic Sans MS'>As calculated t < critical value of t, we may Accept H0 at 0.05 level of significance. Therefore we may conclude that the data are consistent with the assumption of mean I.Q. of 100 in the population.
