# <u>Mann Whitney U Test</u> (Wilcoxon Sum Rank Test) <u>using scipy

***
**A new approach to prenatal care is proposed for pregnant women living in a rural community. The new program involves in-home visits during the course of pregnancy in addition to the usual or regularly scheduled visits. A pilot randomized trial with 15 pregnant women is designed to evaluate whether women who participate in the program deliver healthier babies than women receiving usual care. The outcome is the APGAR scoretext annotation indicator measured 5 minutes after birth. Recall that APGAR scores range from 0 to 10 with scores of 7 or higher considered normal (healthy), 4-6 low and 0-3 critically low. The data are shown below.**

**Is there statistical evidence of a difference in APGAR scores in women receiving the new and enhanced versus usual prenatal care? We run the test using the five-step approach.**

**`Step 1. State the hypotheses. `**

<font face='Comic Sans MS'>Null hypothesis H0: The two populations are equal versus

<font face='Comic Sans MS'>Alternate hypothesis H1: The two populations are not equal.(two-tailed)

**`Step 2. Perform Mann Whitney U Test`**

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
usual_care=[8,7,6,2,5,8,7,3]
new_program=[9,9,7,8,10,9,6]
```

```python
import matplotlib.pyplot as plt
plt.plot(usual_care,marker='o')
plt.plot(new_program,marker='o')
plt.legend(['usual care','new program'])
plt.show()
```

```python
calculated_value,p_value=stats.mannwhitneyu(usual_care,new_program, alternative='two-sided')
```

```python
calculated_value
```

```python
critical_value=10
```

```python
p_value
```

**`Step 3. Conclusion.`**

<font face='Comic Sans MS'>As calculated value < critical value , we may reject the null hypothesis H0. Therefore, we conclude that the populations of APGAR scores are not equal in women receiving usual prenatal care as compared to the new program of prenatal care. 
                                             

<font face='Comic Sans MS' align='center'>or

<font face='Comic Sans MS'>As p_value < 0.05, we may reject the null hypothesis H0. Therefore, we conclude that the populations of APGAR scores are not equal in women receiving usual prenatal care as compared to the new program of prenatal care.
