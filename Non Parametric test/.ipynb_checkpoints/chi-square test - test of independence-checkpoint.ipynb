# <u>Chi-square test - Test of independence using scipy
 

***
**Two sample polls of votes for two candidates A and B for a public office are taken, one from among the residents of rural areas. The results are given in the table. Examine whether the nature of the area is related to voting preference in this election.**



**`Step 1. State the hypotheses.`**

<font face='Comic Sans MS'>Null hypothesis (H0) : the nature of the area is independent of the voting preference in the election.

<font face='Comic Sans MS'>Alternative hypothesis (H1) : the nature of the area is dependent of the voting preference in the election.

**`Step 2. Perform Chi Square Test`**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency
```

```python
data=[[620,380],[550,450]]
```

```python
sns.heatmap(data,annot=True,xticklabels=['A','B'],yticklabels=['Rural','Urban'],fmt='d')
plt.show()
```

```python
calculated_value, p_value, deg_of_freedom, expected=chi2_contingency(data)
```

```python
calculated_value
```

```python
expected
```

```python
critical_value=3.841
```

```python
p_value
```

**`Step 3. Conclusion`**

<font face='Comic Sans MS'>As calculated_value > critical_value, we may reject the H0. Therefore we conclude that nature of the area is dependent of the voting preference in the election.
    
<font face='Comic Sans MS'>or

<font face='Comic Sans MS'>As p-value < 0.05, we may reject the H0. Therefore we conclude that nature of the area is dependent of the voting preference in the election.

**Let's say you want to know if gender has anything to do with political party preference. You poll 440 voters in a simple random sample to find out which political party they prefer. The results of the survey are shown in the table below:**


**To see if gender is linked to political party preference, perform a Chi-Square test of independence.**

**`Step 1. State the hypotheses.`**

<font face='Comic Sans MS'>Null hypothesis (H0): there is no link between gender and political party preference.

<font face='Comic Sans MS'>Alternative hypothesis (H1): there is a link between gender and political party preference.

**`Step 2. Perform Chi Square Test`**

```python
data=[[100,70,30],[140,60,20]]
```

```python
sns.heatmap(data,annot=True,xticklabels=['Republican','Democrat','Independent'],yticklabels=['Male','Female'],fmt='d')
plt.show()
```

```python
calculated_value, p_value, deg_of_freedom, expected=chi2_contingency(data)
```

```python
expected
```

```python
calculated_value
```

```python
critical_value=5.99
```

```python
p_value
```

**`Step 3. Conclusion`**

<font face='Comic Sans MS'>As calculated_value > critical_value, we may reject the H0. Therefore we conclude that there is a link between gender and political party preference.
    
<font face='Comic Sans MS'>or

<font face='Comic Sans MS'>As p-value < 0.05, we may reject the H0. Therefore we conclude that there is a link between gender and political party preference.
